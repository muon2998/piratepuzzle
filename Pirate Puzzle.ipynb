```python
string = "X NET DNA EERHT TINGPORT KEY SRATSEHTKELTTTTLEP THREE LIILENOJNELTTIER FLAG THREE LHFOROLOCEHTX"
string = string.replace(" ","")
print(f"{string} is {len(string)} characters long.")
reverse_string = string[::-1]
print(f"Reverse string is {reverse_string}")
```

```python
divisors = []
for i in range(1,len(string)+1):
    if len(string) % i == 0 and i not in [1,len(string)]:
        divisors.append(i)

print(f"Divisors of {len(string)} are {divisors}")
```

```python
def grid_maker(str,k):
    start_index = 0
    end_index = k
    for position in range(start_index,end_index):
        lst = []
        subset = str[position::k]
        # print(subset)
        for letter in subset:
            lst.append(letter)
        print(' '.join(lst))

for divisor in divisors:
    print(f"\nTrying a {divisor} by {int(len(string)/divisor)} grid:")
    grid_maker(string,divisor)
```
